```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/owid-covid-data.csv', encoding='utf-8')
saudi = data[data['location'] == 'Saudi Arabia']
print(saudi.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(saudi['date'], saudi['new_cases'], color='blue')
plt.title("COVID-19 Daily Cases in Saudi Arabia")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
